1. Question 1:
Create a variable x with the value 15 . Perform floored division by 4 and
print the result. What is the difference between floored division and normal
division?


```python
X = 15
result= X//4

print("floored division result:",result)

#Normal division (/) → gives the exact result as a float.

#Example: 15 / 4 = 3.75

#Floored division (//) → gives the result rounded down to the nearest integer.

#Example: 15 // 4 = 3
```
